```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/WEI_2000_2023_SDI_SubunitScale original.xlsx')

df.head()
df.info()
```

```python
# Select important columns
df = df[['countryName_international',
         'phenomenonTimePeriod_year',
         'phenomenonTimePeriod_quarter',
         'totalWaterAbstraction',
         'netWaterConsumption',
         'renewableWaterResources',
         'WEI+ (%)']]
```

```python
# Rename columns
df.columns = ['Country', 'Year', 'Quarter',
              'Water_Abstraction',
              'Water_Consumption',
              'Renewable_Water',
              'WEI']
```

```python
# Handle missing values
df = df.dropna()
```

```python
# Check duplicate rows
df.duplicated().sum()
```

```python
# Convert data types
df['Year'] = df['Year'].astype(int)
df['WEI'] = pd.to_numeric(df['WEI'], errors='coerce')
```

```python
# Remove any remaining nulls after conversion
df = df.dropna()
```

```python
# Check cleaned data
df.info()
df.head()
```

```python
# Save cleaned dataset
df.to_csv('cleaned_wei_data.csv', index=False)
```
